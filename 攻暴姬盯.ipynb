```python
stat_components = {
    'att_base': ['人物基础', '武器基础'],
    'att_bonus_pct': ['武器%', '套装%', '沙漏%', 'buff%'],
    'att_bonus_flat': ['羽毛白', 'buff白'],
    'crit_rate': ['人物暴%', '武器暴%', '套装暴%', 'buff暴%'],
    'crit_dmg': ['人物爆%', '武器爆%', 'buff爆%'],
}
stat_names = list(stat_components.keys())
```

```python
def dmg(x, bonus=None):
    x_stats = {
        stat_name: sum([
            x[stat_component] for stat_component in stat_components[stat_name]
        ]) for stat_name in stat_names
    }
    if bonus is None:
        bonus = [0, 0, 0]
    bonus_stats = {
        'att_bonus_pct': bonus[0],
        'crit_rate': bonus[1],
        'crit_dmg': bonus[2],
    }

    att = x_stats['att_base'] * (x_stats['att_bonus_pct'] + bonus_stats['att_bonus_pct']) + x_stats['att_bonus_flat']
    crit_rate = x_stats['crit_rate'] + bonus_stats['crit_rate']
    crit_dmg = x_stats['crit_dmg'] + bonus_stats['crit_dmg']

    return att * (1 + crit_rate * crit_dmg)
```

```python
有效词条数 = 32
# 每个有效词条有同等概率是70%, 80%, 90%, 100%, 取平均值85%
# 头因为可以在攻%/暴%/爆%中间选, 故额外计进8词条. 后面计算将会排除.
实际有效词条数 = 有效词条数 * .85 + 8
词条最大值 = np.array([0.058, 0.039, 0.078])
初始估计值 = np.array([实际有效词条数 / 3] * 3)
def 转回有效词条(x):
    res = []
    if x[0] >= 8:
        res.append("攻击头, 攻击%: {:.2f}, 暴击%: {:.2f}, 爆伤%: {:.2f}".format((x[0] - 8)/.85, x[1]/.85, x[2]/.85))
    if x[1] >= 8:
        res.append("暴击头, 攻击%: {:.2f}, 暴击%: {:.2f}, 爆伤%: {:.2f}".format(x[0]/.85, (x[1] - 8)/.85, x[2]/.85))
    if x[2] >= 8:
        res.append("爆伤头, 攻击%: {:.2f}, 暴击%: {:.2f}, 爆伤%: {:.2f}".format(x[0]/.85, x[1]/.85, (x[2] - 8)/.85))
    return res
```

```python
for i in range(len(属性)):
    人物 = 属性.iloc[i]
    
    res = optimize.minimize(
        fun=lambda 词条数: -dmg(人物, 词条数 * 词条最大值),
        x0=初始估计值,
        method='trust-constr',
        bounds = [(0, 实际有效词条数), (0, 实际有效词条数), (0, 实际有效词条数)],
        constraints=[optimize.LinearConstraint(np.array([[1, 1, 1]]), lb=-np.inf, ub=实际有效词条数)],
    )
    print(人物['角色'])
    for s in 转回有效词条(res['x']):
        print('\t{}'.format(s))
```
